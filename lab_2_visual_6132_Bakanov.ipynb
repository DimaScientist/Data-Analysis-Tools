Среди графиков должен присутствовать один boxplot (обязательно надо использовать параметр by, для исследования взаимосвязи числовой переменной и какой-либо другой).
```python
import seaborn as sns

sns.set(rc={'figure.figsize':(12.7,9.27)}) # задаем настройки размера изображений
```

```python
# построим pairplot для следующих признаков: Age, Rating, Upper Salary, Lower Salary, Avg Salary(K)
sns.pairplot(ds_salary_df[["Age", "Rating", "Upper Salary", "Lower Salary", "Avg Salary(K)"]])
```

```python
# зависимость средней зарплаты от рейтинга компании
sns.jointplot(data=ds_salary_df, x="Rating", y="Avg Salary(K)")
```

```python
# распределение зарплаты в зависимсоти от размера компании
sns.violinplot(x="Size", y="Avg Salary(K)", data=ds_salary_df)
```

```python
# отберем числовые колонки
numeric_types = [np.int8, np.int16, np.int32, np.int64, np.float16, np.float32, np.float64]
numeric_columns = [col for col in ds_salary_df if ds_salary_df[col].dtype in numeric_types]
numeric_columns
```

```python
# построим тепловую карту, показывающую корреляцию между признаками
sns.heatmap(ds_salary_df[numeric_columns].corr())
```

*4 Построить QQ-график (и гистограмму) для одного из числовых признаков с целью проверки на соответствие его распределения нормальному распределению*

```python
from scipy.stats import probplot # импортируем QQ-график из scipy.stats
```

```python
# гистограмма для средней зарплаты в компании
ds_salary_df["Avg Salary(K)"].hist()
```

```python
# квантиль-квантильный график для Avg Salary(K)
# на проверку соответствия нормальному распределению
fig, ax = plt.subplots(figsize=(10, 10))
probplot(ds_salary_df["Avg Salary(K)"], plot=ax)
plt.show()
```
